```python
from scipy import constants ### biblioteca de constantes
import matplotlib.pyplot as plt
import numpy as np

print(1e9)
### Importacion de las constantes ###
c = constants.speed_of_light
h = constants.h
k = constants.k

### definicion de las constantes de la funcion
C1 = 2.0*h*c**2*np.pi
C2 = h*c/k

def funcion_planck(lambd, T, C1, C2):
    intensidad = C1 / ( (lambd**5) * (np.exp(C2/(lambd*T)) - 1.0) )
    return intensidad

temperatura = [100,5800,120000]

#Maximos de lambda segun la ley de desplazamiento de Wien
LambdasMax=np.zeros(np.size(temperatura))
for i in range(len(T)):
    LambdasMax[i]=2.897777291*10**(-3)/temperatura[i] # en metros
print(LambdasMax)

plt.figure()
lamb = np.linspace(0.5e-7, 1.5e-4, 1000)
plt.plot(lamb*1e9, funcion_planck(lamb, 100, C1, C2)*1e-9)
plt.scatter(LambdasMax[0]*1e9, funcion_planck(LambdasMax[0],100,C1,C2)*1e-9,color="red",marker="+",label=r"I max; $\lambda$="+str(round(LambdasMax[0],6)*1e9)+" nm",s=100)
plt.grid('..')
plt.xlabel("Longitud de onda "+r"$\lambda$ "+"en nm")
plt.ylabel("Intensidad en W.m-2.sr-1.nm-1")
plt.title("Ley de Planck para T=100 K")
plt.legend()
plt.savefig('Ley_planck_100.png', dpi=100, facecolor='white')
plt.show()

plt.figure()
lamb = np.linspace(0.1e-8, 0.3e-5, 1000)
plt.plot(lamb*1e9, funcion_planck(lamb, 5800, C1, C2)*1e-9)
plt.scatter(LambdasMax[1]*1e9, funcion_planck(LambdasMax[1],5800,C1,C2)*1e-9,color="red",marker="+",label=r"I max; $\lambda$="+str(round(LambdasMax[1],7)*1e9)+" nm", s=100)
plt.grid('..')
plt.title("Ley de Planck para T=5800 K")
plt.xlabel("Longitud de onda "+r"$\lambda$ "+"en nm")
plt.legend()
plt.ylabel("Intensidad en W.m-2.sr-1.nm-1")
plt.savefig('Ley_planck_5800.png', dpi=100, facecolor='white')
plt.show()

plt.figure()
lamb = np.linspace(1e-10, 1e-7, 1000)
plt.plot(lamb*1e9, funcion_planck(lamb, 120000, C1, C2)*1e-9)
plt.scatter(LambdasMax[2]*1e9, funcion_planck(LambdasMax[2],120000,C1,C2)*1e-9,color="red",marker="+",label="I max; "+r"$\lambda$"+"="+str(round(LambdasMax[2],9)*1e9)+" nm", s=100)
plt.legend()
plt.grid('..')
plt.title("Ley de Planck para T=120000 K")
plt.xlabel("Longitud de onda "+r"$\lambda$ "+"en nm")
plt.ylabel("Intensidad en W.m-2.sr-1.nm-1")
plt.savefig('Ley_planck_120000.png', dpi=100, facecolor='white')
plt.show()

print(LambdasMax[2])
```
